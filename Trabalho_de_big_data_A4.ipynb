## **Dataframe**
Importamos as blibiotecas necessárias, pandas e ploty, lemos o dataset e então, fizemos o dataframe:


```python
#Importação de bibliotecas

import plotly.express as px
import pandas as pd
path = r"/content/StudentsPerformance.csv"
path
```

```python
tabela_total = pd.DataFrame()
tabela = pd.read_csv(path)
tabela_total = tabela_total.append(tabela)
display(tabela_total)
```

## **Filtragem de colunas**
Filtramos a coluna que mostra a escolariedade dos pais e criamos outra coluna com a  somatória dos filhos de cada nível de eduação. Assim, podemos ver melhor a influência que a escolariedade tem sobre esse aspecto da vida.

```python
# Contando a quantidade de alunos, por nivel de educação dos pais
tabela_parental = tabela_total.filter(items=["parental level of education", "lunch"])
tabela_parental.rename(columns={"lunch": "quantidade"}, inplace=True)
tabela_parental = tabela_parental.groupby("parental level of education").count().reset_index()
tabela_parental
```

## **Gráfico**
Com base na filtragem de colunas acima, criamos o gráfico a seguir:

```python
#Criando um grafico da quantidade de cada nível de educação dos pais dos alunos
grafico = px.bar(tabela_parental, x="parental level of education", y="quantidade", title="Quantidade de alunos por escolaridade dos pais")
grafico.show()
```

## **Média das 3 notas**
Cálculo da média das notas de matemática, leitura e redação de cada aluno:

```python
tabela_notas = tabela_total.filter(items=["math score", "reading score", "writing score"])
tabela_notas
```

```python
print("A media das notas de matematica: {}".format(tabela_notas["math score"].mean()))
print("A media das notas de leitura: {}".format(tabela_notas["reading score"].mean()))
print("A media das notas de redacao: {}".format(tabela_notas["writing score"].mean()))
```

## **Cálculo de Indicadores**
A seguir mostra o cálculo de três indicadores, o das maiores notas de matemática, leitura e redação:

```python
#As maiores notas de Matemática
print("10 maiores notas de matematica:")
tabela_total["math score"].nlargest(10)
```

```python
#As maiores notas de Leitura
print("10 maiores notas de leitura:")
tabela_total["reading score"].nlargest(10)
```
